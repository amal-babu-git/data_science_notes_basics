```python

data_cleaned=data_cleaned.drop(['Price'],axis=1)
```

```python
data_cleaned.head()
```

```python
#the next OLS assumption is multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

for find multicollinearity , use VIF --> Varance inflation factor
0 means no multicollinearity
<=5 is good 
>5 or morethan 10 ,p value is qustanable,should remove any data for fix it

```python
variables=data_cleaned[['Mileage','Year','EngineV']]
vif=pd.DataFrame()
vif["VIF"]=[variance_inflation_factor(variables.values,i) for i in range(variables.shape[1])]#shape[1] means number of colums,[0]means rows
vif['Features']=variables.columns
vif
```

```python
data_no_multicollinearity=data_cleaned.drop(['Year'],axis=1)#fixing multicollinarity by removing year data from data_fame
data_no_multicollinearity
```

```python
#creating dummy for catagoriacl variables
data__with_dummies=pd.get_dummies(data_no_multicollinearity,drop_first=True)
data__with_dummies.head()
```

```python
#rearranging data
data__with_dummies.columns.values
```

```python
cols=['log_price','Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
       
```

```python
data_preprocessed=data__with_dummies[cols]
data_preprocessed.head()
```

```python
#LinearRegression
#declaring target and inputs
target=data_preprocessed['log_price']
inputs=data_preprocessed.drop(['log_price'],axis=1)
```

```python
#scal the data --> x-u/sigma
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(inputs)
input_scaled=scaler.transform(inputs)
```

```python
#train set split
from sklearn.model_selection import train_test_split

x_train , x_test,y_train,y_test=train_test_split(input_scaled,target,test_size=0.2,random_state=365)
```

```python
#regression
reg=LinearRegression()
reg.fit(x_train,y_train)
y_hat=reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel("Targets(y_train)",size=18)
plt.ylabel("Predictions(y_hat)",size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.displot(y_train-y_hat)
plt.title("Rasidual PDF",size=18)
#residuals are diff b/w target and predicctions
#errors must be normally sistributed with mean 0 ,
# The residuals are the estimates of errors
```

```python
reg.intercept_ 
#Bias
```

```python
reg.coef_
#Weights
```

Crating new dataframe for analysing weights

Weights interprestation
   1)Countinuous variables--> +ve value shows that feture increases in values,and price respectivily
                          --> -ve value shows that feature increases in values, log-price and price decreases
   2)Dummy variable       -->A +ve wight shows that the respective catagory(Brand) is more expensive  than the bench mark(Audi)
                          -->A -ve values shows that the respective catagory is less expensive than the bench mark

```python
reg_summary=pd.DataFrame()
reg_summary['Features']=inputs.columns.values
reg_summary['Weights']=reg.coef_
reg_summary
```

```python
#Test
y_hat_test=reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test,alpha=0.2)
plt.xlabel("Target(y_test)",size=18)
plt.ylabel("Predictions(y_hat_test)",size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
#dataframe perfomance
df_pf=pd.DataFrame(np.exp(y_hat_test),columns=['Predictions'])#change log predictions to actual value using np.exp()
df_pf
```

```python
df_pf['Target']=np.exp(y_test)
df_pf
```
